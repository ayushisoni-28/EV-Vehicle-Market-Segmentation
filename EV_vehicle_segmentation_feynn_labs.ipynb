```python

import matplotlib.pyplot as plt

# Adjust column names for better readability
data2.columns = [
    'S No',
    'Vehicle Class',
    'ELECTRIC(BOV)',
    'FUEL CELL HYDROGEN',
    'PLUG-IN HYBRID EV',
    'PURE EV',
    'SOLAR',
    'STRONG HYBRID EV',
    'TOTAL'
]

# Set 'Vehicle Class' as the index for easier plotting
data2.set_index('Vehicle Class', inplace=True)

# Extracting fuel-related columns
fuel_columns = [
    'ELECTRIC(BOV)',
    'FUEL CELL HYDROGEN',
    'PLUG-IN HYBRID EV',
    'PURE EV',
    'SOLAR',
    'STRONG HYBRID EV'
]

# Transpose data to have fuels on the x-axis and vehicle classes as separate bars
data_transposed = data2[fuel_columns].T

# Plot grouped bar chart
data_transposed.plot(kind='bar', figsize=(12, 8), width=0.8)

# Chart labels and title
plt.title('Vehicle Categories by Fuel Type', fontsize=16)
plt.xlabel('Fuel Type', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Vehicle Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt




# Sort the data by the 'TOTAL' column in descending order and select the top 5
top_5_data = data2.sort_values('TOTAL', ascending=False).head(5)

# Extracting fuel-related columns for the top 10 vehicle categories
fuel_columns = [
    'ELECTRIC(BOV)',
    'FUEL CELL HYDROGEN',
    'PLUG-IN HYBRID EV',
    'PURE EV',
    'SOLAR',
    'STRONG HYBRID EV'
]

# Transpose data to have fuels on the x-axis and vehicle classes as separate bars
data_transposed = top_5_data[fuel_columns].T

# Plot grouped bar chart
data_transposed.plot(kind='bar', figsize=(12, 8), width=0.8)

# Chart labels and title
plt.title('Top 5 Vehicle Categories by Fuel Type', fontsize=16)
plt.xlabel('Fuel Type', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Vehicle Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# List of fuel types to plot
fuel_columns = [
    'ELECTRIC(BOV)',
    'FUEL CELL HYDROGEN',
    'PLUG-IN HYBRID EV',
    'PURE EV',
    'SOLAR',
    'STRONG HYBRID EV'
]

# Plot individual bar charts for each fuel type
for fuel in fuel_columns:
    plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
    data2[fuel].plot(kind='bar', color='skyblue', edgecolor='black')

    # Add titles and labels
    plt.title(f'Number of Vehicles for {fuel}', fontsize=16)
    plt.xlabel('Vehicle Category', fontsize=12)
    plt.ylabel('Number of Vehicles', fontsize=12)

    # Adjust x-axis label rotation and alignment
    plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate and align labels to the right
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Adjust layout for better spacing
    plt.tight_layout()
    plt.show()
```

```python
# Sum up total counts for each fuel type
fuel_totals = data2[fuel_columns].sum()

# Plot pie chart with percentages outside the chart
fig, ax = plt.subplots(figsize=(8, 8))
pie = ax.pie(
    fuel_totals,
    autopct='%1.1f%%',  # Show percentage with one decimal
    startangle=90,
    colors=plt.cm.Paired.colors,  # Use a colormap
    wedgeprops={'edgecolor': 'black'},  # Add black border to the slices
    pctdistance=1.05,  # Move percentage text outside (default is 0.6)
    labeldistance=1.1  # Move labels outside the pie chart (default is 1.05)
)

# Title and other adjustments
plt.title('Market Share of Fuel Types', fontsize=16)
plt.ylabel('')  # Remove y-axis label for a cleaner look

# Create a legend with the fuel categories and corresponding colors
labels = fuel_columns  # These are the fuel types to show in the legend
ax.legend(labels=labels, loc='upper left', bbox_to_anchor=(1.05, 1), title="Fuel Types", fontsize=10)

# Ensure the plot does not cut off and has proper spacing
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Sum up total counts for each fuel type
fuel_totals = data2[fuel_columns].sum()

# Calculate the total sum of all fuel types
total_count = fuel_totals.sum()

# Calculate the percentage of each fuel type
fuel_percentages = (fuel_totals / total_count) * 100

# Print the percentages for each fuel type
for fuel, percentage in fuel_percentages.items():
    print(f"{fuel}: {percentage:.1f}%")
```

```python
import seaborn as sns

# Calculate correlation matrix
fuel_corr = data2[fuel_columns].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(fuel_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Fuel Types', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Clean up the column names
data1.columns = [
    'S No',
    'Vehicle Class',
    'ELECTRIC(BOV)',
    'FUEL CELL HYDROGEN',
    'PLUG-IN HYBRID EV',
    'PURE EV',
    'SOLAR',
    'STRONG HYBRID EV',
    'TOTAL'
]

# Set 'Vehicle Class' as the index for easier plotting
data1.set_index('Vehicle Class', inplace=True)

# Extract the top 10 vehicle classes by 'TOTAL' column
top_10_classes = data1['TOTAL'].nlargest(10)

# Plot the bar chart
fig, ax = plt.subplots(figsize=(12, 8))

# Bar chart of the top 10 vehicle classes across fuel types
data1.loc[top_10_classes.index, fuel_columns].plot(kind='bar', ax=ax, width=0.8)

# Title and labels
plt.title('Top 10 Vehicle Classes by Fuel Type', fontsize=16)
plt.xlabel('Vehicle Class', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right', fontsize=10)

# Add legend
plt.legend(title="Fuel Types", bbox_to_anchor=(1.05, 1), loc='upper left')

# Optional: Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ensure the plot does not cut off and has proper spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix for fuels
fuel_corr = data1[fuel_columns].corr()

# Plot correlation matrix for fuels
plt.figure(figsize=(10, 8))
sns.heatmap(fuel_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)

# Title and display
plt.title('Correlation Matrix for Fuels', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the total number of vehicles for 'ELECTRIC(BOV)' fuel type
electric_bov_totals = data1['ELECTRIC(BOV)']

# Sort by the total number of vehicles and get the top 5
top_5_electric_bov = electric_bov_totals.nlargest(5)

# Plot pie chart for the top 5 vehicle classes
plt.figure(figsize=(10, 8))
pie = plt.pie(top_5_electric_bov, autopct='%1.1f%%', startangle=90,
              colors=plt.cm.Paired.colors, wedgeprops={'edgecolor': 'black'},
              textprops={'fontsize': 10})

# Title and adjustments
plt.title('Top 5 Vehicle Classes Segmentation by ELECTRIC(BOV) Fuel', fontsize=16)
plt.ylabel('')  # Remove the y-label for a cleaner look

# Add legend outside the pie chart
plt.legend(top_5_electric_bov.index, title="Vehicle Class", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# List of fuel types to plot
fuel_types = [
    'FUEL CELL HYDROGEN',
    'PLUG-IN HYBRID EV',
    'PURE EV',
    'SOLAR',
    'STRONG HYBRID EV'
]

# Loop over each fuel type and create a pie chart
for fuel in fuel_types:
    # Calculate the total number of vehicles for the current fuel type
    fuel_totals = data1[fuel]

    # Sort by the total number of vehicles and get the top 5
    top_5_fuel = fuel_totals.nlargest(5)

    # Calculate the percentage for each fuel type
    total_top_5 = top_5_fuel.sum()
    percentage_top_5 = (top_5_fuel / total_top_5) * 100

    # Print the results as percentages
    print(f"\nPercentage Distribution of {fuel}:")
    for vehicle_class, percentage in percentage_top_5.items():
        print(f"{vehicle_class}: {percentage:.2f}%")

    # Plot pie chart for the top 5 vehicle classes for the current fuel type
    plt.figure(figsize=(10, 8))
    pie = plt.pie(top_5_fuel, autopct='%1.1f%%', startangle=90,
                  colors=plt.cm.Paired.colors, wedgeprops={'edgecolor': 'black'},
                  textprops={'fontsize': 10})

    # Title and adjustments
    plt.title(f'Top 5 Vehicle Classes Segmentation by {fuel} Fuel', fontsize=16)
    plt.ylabel('')  # Remove the y-label for a cleaner look

    # Add legend outside the pie chart
    plt.legend(top_5_fuel.index, title="Vehicle Class", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

    # Ensure the plot is displayed correctly
    plt.tight_layout()
    plt.show()
```
